## Line Chart

```python
# download dataset
!wget https://raw.githubusercontent.com/Njmondoa/Superstore-Sales-Analysis/main/SuperStore_data.csv
```

```python
import pandas as pd
df = pd.read_csv('SuperStore_data.csv')
```

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format = 'mixed')
```

```python
#Set 'Order Date' as index
df = df.set_index('Order Date')
```

```python
# Resample by Month and sum over features
monthly_df = df.resample('M').sum()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(10,4))

# Sales plot
ax[0].plot(monthly_df.index, monthly_df['Sales'], color = 'blue')
ax[0].set_title('Monthly Sale')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Sales')
ax[0].tick_params(axis='x', rotation=45)

# Profit plot
ax[1].plot(monthly_df.index, monthly_df['Profit'], color = 'red')
ax[1].set_title('Monthly Profit')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Profit')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
yearly_df = df.resample('Y').sum()
```

```python
fig , ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,4))

# Sales plot
ax[0].plot(yearly_df.index, yearly_df['Sales'], color = 'blue')
ax[0].set_title('yearly Sale')
ax[0].set_xlabel('year')
ax[0].set_ylabel('Sales')
ax[0].tick_params(axis='x', rotation=45)

# Profit plot
ax[1].plot(yearly_df.index, yearly_df['Profit'], color = 'red')
ax[1].set_title('yearly Profit')
ax[1].set_xlabel('year')
ax[1].set_ylabel('Profit')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
